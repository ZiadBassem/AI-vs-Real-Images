```python
import os, torch, zipfile
from tqdm import tqdm
from PIL import Image
from diffusers import StableDiffusionPipeline
```

```python
# Make sure GPU is enabled (Runtime → Change runtime type → GPU)
device = "cuda" if torch.cuda.is_available() else "cpu"
print("✅ Using device:", device)

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    dtype=torch.float16 if device == "cuda" else torch.float32
).to(device)
```

```python
categories = {
    "sweater_weather": {
        "prompt": "A cozy autumn scene, someone wearing a warm knitted sweater, fall leaves, cinematic lighting, ultra realistic photography",
        "count": 100
    },
    "night": {
        "prompt": "A realistic night time city scene, glowing neon lights, starry sky, dark atmosphere, ultra realistic photography",
        "count": 100
    },
    "nature": {
        "prompt": "A scenic nature landscape, lush forest, mountains, sunlight, vibrant colors, ultra realistic photography",
        "count": 100
    },
    "underwater": {
        "prompt": "A beautiful underwater scene, coral reefs, tropical fishes, clear ocean water, underwater photography, vibrant colors",
        "count": 100
    },
    "vintage": {
        "prompt": "A vintage style photograph, sepia tones, retro aesthetics, old film camera effect, classic photography",
        "count": 50
    },
    "pumpkin_spice": {
        "prompt": "A cozy coffee shop scene with a pumpkin spice latte, fall decor, warm tones, autumn atmosphere, ultra realistic photography",
        "count": 50
    }
}

output_dir = "ai_dataset"
os.makedirs(output_dir, exist_ok=True)
```

```python
BATCH_SIZE = 4
TARGET_SIZE = (224,224)

for label, info in categories.items():
    class_dir = os.path.join(output_dir, label)
    os.makedirs(class_dir, exist_ok=True)

    print(f"\n✅ Generating {info['count']} images for: {label}")
    total_batches = (info["count"] + BATCH_SIZE - 1) // BATCH_SIZE
    img_idx = 0

    for b in tqdm(range(total_batches)):
        num_to_generate = min(BATCH_SIZE, info["count"] - img_idx)
        images = pipe(
            info["prompt"],
            num_inference_steps=20,
            num_images_per_prompt=num_to_generate
        ).images

        for img in images:
            img = img.resize(TARGET_SIZE, Image.Resampling.LANCZOS)
            img.save(os.path.join(class_dir, f"{label}_{img_idx}.png"))
            img_idx += 1

print("\n🎉 Done! Dataset created at 224x224 under ai_dataset/")
```

```python
zip_name = "ai_dataset.zip"
with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(output_dir):
        for file in files:
            file_path = os.path.join(root, file)
            zipf.write(file_path, os.path.relpath(file_path, output_dir))

print(f"📦 Dataset zipped as {zip_name}")
```
